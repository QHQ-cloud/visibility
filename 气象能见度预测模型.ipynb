```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as LR
from sklearn.preprocessing import PolynomialFeatures as pnf
from sklearn.metrics import r2_score,mean_squared_error as mse
```

```python
X_data = newdata.iloc[:,0:13]
```

```python
Y_data = newdata.iloc[:,-1]
```

```python
Y_data.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_data,Y_data,test_size = 0.3,random_state = 420)  # 划分训练集和测试集
```

```python
x_train = x_train.reset_index(drop = True)
x_test = x_test.reset_index(drop = True)
y_train = y_train.reset_index(drop = True)
y_test = y_test.reset_index(drop = True)# 因为随机划分 序列混乱 必须重置索引 
```

```python
x_train  
```

```python
x_test.head()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_test.head()
```

```python
# from sklearn.model_selection import train_test_split
# from sklearn.linear_model import LinearRegression as LR
# from sklearn.preprocessing import PolynomialFeatures as pnf
# from sklearn.metrics import r2_score,mean_squared_error as mse
model = LR().fit(x_train,y_train)  # 建立线性模型
```

```python
y_predict = model.predict(x_train)  # 这是把训练集的自变量输入查看预测结果
```

```python
r2_score(y_train,y_predict) # 查看R平方
```

```python
sorted([*zip(x_train.columns,model.coef_/1e14)],key = lambda x:x[1],reverse=True)



#######注意这里 新添加的model（单纯线性回归的系数）
```

```python
model.intercept_ / 1e14   #  截距 


#############################   这是截距 
```

```python
from sklearn.ensemble import RandomForestRegressor as rfr  # 随机森林模型 非线性回归手段
model_rfr = rfr(n_estimators=100,random_state=420).fit(x_train,y_train)
```

```python
predict_rfr_train = model_rfr.predict(x_train)
```

```python
r2_score(y_train,predict_rfr_train)
```

```python
mse(y_train,predict_rfr_train)  # 因为结果较大，mse 又没有上限 因此参考意义不大 R平方可以拟合 基本趋势
```

```python
predict_rfr_test = model_rfr.predict(x_test)
```

```python
r2_score(y_test,predict_rfr_test)  # 也米有过拟合现象 （百度搜索过拟合）
```
